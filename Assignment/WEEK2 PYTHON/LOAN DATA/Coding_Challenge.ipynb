## 8. Apply Function

```python
# if year greater than 2019 it is termed as COVID
def fun(value):
    if value>2019:
        return "COVID"
    else:
        return "No"

data_copy=data.copy()
data_copy['Year_mod'] = data_copy['Year'].apply(fun)
data_copy.head(-10)
```

## 9. By using the lambda operator

```python
# adding 10 years to the existing year
data_copy = data.copy()
data_copy['Year'] = data_copy['Year'].apply(lambda x: x + 10)
display(data_copy)
```
